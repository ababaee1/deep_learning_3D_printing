```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from PIL import Image,ImageOps
from skimage.metrics import structural_similarity, mean_squared_error
import cv2 as cv
import os
```

```python
rootpath="Z:/UBC/MANU465/Project/Single_beam_gcode/Imageset/"
#The list of subfolder, each contains images of a beam
Folderlist = ["RP_B01",
              "RP_B02",
              "RP_B03",
              "RP_B04",
              "RP_B05",
              "RP_B06",
              "RP_B07",
              "RP_B08",
              "RP_B09",
              "RP_B10",
              "RP_B11",
              "RP_B12",
              "RP_B13",
              "RP_B14",
              "RP_B15"]


#Loop through each beam
for Folder in Folderlist:
    path=rootpath+"/"+Folder
    files = [file for file in os.listdir(path) if file.endswith('png')]
    #Open the First image in the folder, create one normal image and one vertically flipped image
    #Since I stack all the image vertically, I only need to horizontally flip them at the end to get the flipped image
    Image_com_Normal = Image.open(path+"/"+files[0])
    Image_com_Vflip = ImageOps.flip(Image_com_Normal)

    #Also we can stack them in the reverse order,check the loop below
    Image_com_Rev_Normal=Image_com_Normal
    Image_com_Rev_Vflip=Image_com_Vflip

    Image_com_Normal = np.array(Image_com_Normal)
    #Flip the order so that is width,height
    #Record the size of the first image and use it to resize all other images
    Imageshape=(Image_com_Normal.shape[1],Image_com_Normal.shape[0])


    #go through all the images in the folder and stack them vertically
    for file in files:
        #Skip the first image
        if file!=files[0]:
            #Import the next image
            Imagenext = Image.open(path+"/"+file)
            #Stack them
            Image_com_Normal=np.vstack([Image_com_Normal,Imagenext.resize(Imageshape)])
            Image_com_Vflip=np.vstack([Image_com_Vflip,ImageOps.flip(Imagenext.resize(Imageshape))])
            #Stack them in reverse order
            Image_com_Rev_Normal=np.vstack([Imagenext.resize(Imageshape),Image_com_Rev_Normal])
            Image_com_Rev_Vflip=np.vstack([ImageOps.flip(Imagenext.resize(Imageshape)),Image_com_Rev_Vflip])

    #For verification
    print(Image_com_Normal.shape)

    #Convert the array back to image for output
    Image_com_Normal=Image.fromarray(Image_com_Normal)
    Image_com_Vflip=Image.fromarray(Image_com_Vflip)
    Image_com_Rev_Normal=Image.fromarray(Image_com_Rev_Normal)
    Image_com_Rev_Vflip=Image.fromarray(Image_com_Rev_Vflip)


    #Create two more photos by flipping them horizontally
    Image_com_Hflip=ImageOps.mirror(Image_com_Normal)
    Image_com_VHflip=ImageOps.mirror(Image_com_Vflip)
    Image_com_Rev_Hflip=ImageOps.mirror(Image_com_Rev_Normal)
    Image_com_Rev_VHflip=ImageOps.mirror(Image_com_Rev_Vflip)

    #Output all the images
    Image_com_Normal.save(Folder+'_01_normal.png' )
    Image_com_Vflip.save(Folder+'_02_Vflip.png' )
    Image_com_Hflip.save(Folder+'_03_Hflip.png' )
    Image_com_VHflip.save(Folder+'_04_VHflip.png' )

    Image_com_Rev_Normal.save(Folder+'_05_Rev_normal.png' )
    Image_com_Rev_Vflip.save(Folder+'_06_Rev_Vflip.png' )
    Image_com_Rev_Hflip.save(Folder+'_07_Rev_Hflip.png' )
    Image_com_Rev_VHflip.save(Folder+'_08_Rev_VHflip.png' )

```
